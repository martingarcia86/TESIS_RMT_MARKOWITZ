```python
# Notebook 03 - Matriz de covarianza muestral

## Objetivo

Estimar las matrices de covarianza y correlación a partir de los rendimientos financieros.

## Entradas

- Matriz de rendimientos logarítmicos.

## Tareas principales

1. Calcular la matriz de covarianza.
2. Calcular la matriz de correlación.
3. Obtener valores y vectores propios.
4. Analizar las propiedades espectrales.

## Salidas

- Matriz de covarianza muestral.
- Matriz de correlación.
- Espectro de valores propios.

## Siguiente paso

Las matrices obtenidas se emplean en el Notebook 04 para aplicar técnicas de limpieza basadas en Teoría de Matrices Aleatorias.





# =========================================================
# Notebook 03:
# Construcción de matriz de covarianza muestral
# =========================================================

# pandas:
# Manejo de tablas y bases de datos
import pandas as pd

# numpy:
# Operaciones numéricas y matriciales
import numpy as np

# pathlib:
# Manejo profesional de rutas
from pathlib import Path


# =========================================================
# Definir rutas del proyecto
# =========================================================

# Ruta principal del proyecto de tesis
BASE_DIR = Path(r"C:\MRGB\Personal\Tesis")

# Carpeta donde se guardan los resultados
OUTPUT_DIR = BASE_DIR / "outputs"

# Carpeta donde están las bases ya procesadas
DATOS_PROCESADOS_DIR = OUTPUT_DIR / "datos_procesados"


# =========================================================
# Cargar base final de rendimientos logarítmicos
# =========================================================

# Archivo generado en el Notebook 02
archivo_rendimientos = DATOS_PROCESADOS_DIR / "rendimientos_logaritmicos_80pct.csv"

# Leer archivo CSV
rendimientos = pd.read_csv(archivo_rendimientos)

# Convertir la columna Fecha a formato fecha
rendimientos["Fecha"] = pd.to_datetime(rendimientos["Fecha"])


# =========================================================
# Separar fechas y matriz numérica de rendimientos
# =========================================================

# Guardamos las fechas aparte porque no entran al cálculo de covarianza
fechas = rendimientos["Fecha"]

# Nos quedamos únicamente con columnas de rendimientos
R = rendimientos.drop(columns=["Fecha"])

# Aseguramos que todas las columnas sean numéricas
R = R.apply(pd.to_numeric, errors="coerce")


# =========================================================
# Verificación de la matriz de rendimientos
# =========================================================

print("Dimensiones de la matriz de rendimientos R:")
print(R.shape)

print("\nColumnas no numéricas:")
print(R.select_dtypes(exclude="number").columns.tolist())

print("\nValores faltantes en R:")
print(R.isna().sum().sum())

display(R.head())


# =========================================================
# Construcción de matriz de covarianza muestral
# =========================================================

# R.cov() calcula la covarianza muestral entre columnas.
# Cada columna representa un activo financiero.
# Cada fila representa una observación diaria de rendimiento.
Sigma_muestral = R.cov()

# =========================================================
# Exportar matriz de covarianza muestral
# =========================================================

# Nombre del archivo de salida
archivo_sigma = DATOS_PROCESADOS_DIR / "sigma_muestral_80pct.csv"

# Guardar CSV
Sigma_muestral.to_csv(archivo_sigma)

# Confirmación
print("Matriz de covarianza exportada correctamente.")
print(f"Ruta: {archivo_sigma}")


# =========================================================
# Verificación de la matriz de covarianza
# =========================================================

print("Dimensiones de la matriz de covarianza muestral:")
print(Sigma_muestral.shape)

display(Sigma_muestral.head())
```
