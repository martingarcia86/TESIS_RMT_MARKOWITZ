```python
# Notebook 02 - Cálculo de rendimientos

## Objetivo

Calcular los rendimientos logarítmicos a partir de las series de precios procesadas.

## Entradas

- Series de precios obtenidas en el Notebook 01.

## Tareas principales

1. Importar las series de precios.
2. Calcular los rendimientos logarítmicos.
3. Eliminar valores faltantes.
4. Exportar la matriz de rendimientos.

## Salidas

- Matriz de rendimientos logarítmicos.

## Siguiente paso

La matriz de rendimientos se utiliza en el Notebook 03 para estimar la matriz de covarianza muestral.




# =========================================================
# Notebook 02: Construcción de rendimientos logarítmicos
# =========================================================

import pandas as pd
import numpy as np
from pathlib import Path

BASE_DIR = Path(r"C:\MRGB\Personal\Tesis")
OUTPUT_DIR = BASE_DIR / "outputs"
DATOS_PROCESADOS_DIR = OUTPUT_DIR / "datos_procesados"

archivo_precios_limpios = DATOS_PROCESADOS_DIR / "precios_limpios_al_80pct.csv"

precios = pd.read_csv(archivo_precios_limpios)

precios["Fecha"] = pd.to_datetime(precios["Fecha"])
precios = precios.sort_values("Fecha")

print(precios.shape)
display(precios.head())

# =========================================
# Construcción de rendimientos logarítmicos
# =========================================

# Crear copia de la base
rendimientos = precios.copy()

# Eliminar columna fecha temporalmente
fechas = rendimientos["Fecha"]

rendimientos = rendimientos.drop(columns=["Fecha"])

# Fórmula:
# r_t = ln(P_t / P_(t-1))

rendimientos = np.log(rendimientos / rendimientos.shift(1))

# Volver a agregar fechas
rendimientos.insert(0, "Fecha", fechas)

# Mostrar dimensiones
print("Dimensiones:")
print(rendimientos.shape)

# Mostrar primeras filas
display(rendimientos.head())

# =========================================
# Análisis de valores faltantes
# en rendimientos
# =========================================

# Contar valores faltantes por columna
faltantes_rend = rendimientos.isna().sum()

# Convertir a DataFrame
faltantes_rend = faltantes_rend.reset_index()

# Renombrar columnas
faltantes_rend.columns = ["Activo", "Valores_Faltantes"]

# Calcular porcentaje faltante
faltantes_rend["Porcentaje_Faltante"] = (
    faltantes_rend["Valores_Faltantes"] / len(rendimientos)
) * 100

# Ordenar de mayor a menor
faltantes_rend = faltantes_rend.sort_values(
    by="Porcentaje_Faltante",
    ascending=False
)

# Mostrar resultados
display(faltantes_rend.head(15))

# =========================================
# Filtrar emisoras con demasiados NaN
# en rendimientos
# =========================================

# Definir umbral máximo permitido
umbral_nan = 20

# Seleccionar emisoras aceptables
activos_validos = faltantes_rend[
    faltantes_rend["Porcentaje_Faltante"] < umbral_nan
]["Activo"]

# Conservar columna fecha
columnas_finales = ["Fecha"] + activos_validos.tolist()

# Crear matriz limpia de rendimientos
rendimientos_limpios = rendimientos[columnas_finales]

# Mostrar dimensiones
print("Dimensiones originales:")
print(rendimientos.shape)

print("\nDimensiones después de limpieza:")
print(rendimientos_limpios.shape)

# Número de activos finales
print("\nNúmero de activos conservados:")
print(rendimientos_limpios.shape[1] - 1)

# Mostrar primeras filas
display(rendimientos_limpios.head())

# =========================================
# Eliminar filas restantes con NaN
# =========================================

# Eliminar filas incompletas
rendimientos_finales = rendimientos_limpios.dropna()

# Reiniciar índice
rendimientos_finales = rendimientos_finales.reset_index(drop=True)

# Mostrar dimensiones finales
print("Dimensiones finales:")
print(rendimientos_finales.shape)

# Verificar si quedan NaN
print("\nValores faltantes restantes:")
print(rendimientos_finales.isna().sum().sum())

# Mostrar primeras filas
display(rendimientos_finales.head())

# =========================================================
# Guardar rendimientos finales
# =========================================================

archivo_rendimientos = DATOS_PROCESADOS_DIR / "rendimientos_logaritmicos_80pct.csv"

rendimientos_finales.to_csv(archivo_rendimientos, index=False)

print("Rendimientos guardados correctamente.")
print(f"Ruta de salida: {archivo_rendimientos}")
```
