```python
# Notebook 05 - Optimización de portafolios

## Objetivo

Construir portafolios de mínima varianza y comparar los resultados obtenidos con matrices de covarianza originales y matrices regularizadas.

## Entradas

- Matriz de covarianza muestral.
- Matriz de covarianza obtenida mediante Teoría de Matrices Aleatorias.

## Tareas principales

1. Calcular los pesos del portafolio de mínima varianza.
2. Evaluar el riesgo asociado.
3. Comparar los resultados obtenidos con ambas matrices.
4. Analizar el efecto de la limpieza espectral.

## Salidas

- Pesos óptimos.
- Medidas de riesgo.
- Tablas y gráficas comparativas.

## Resultado final

Evaluar el impacto de la Teoría de Matrices Aleatorias sobre la optimización de portafolios.





# =========================================================
# Notebook 05:
# Optimización de portafolios de Markowitz
# =========================================================

# pandas:
# Manejo de tablas
import pandas as pd

# numpy:
# Operaciones matriciales
import numpy as np

# matplotlib:
# Gráficas
import matplotlib.pyplot as plt

# pathlib:
# Manejo de rutas
from pathlib import Path


# =========================================================
# Definir rutas del proyecto
# =========================================================

BASE_DIR = Path(r"C:\MRGB\Personal\Tesis")

OUTPUT_DIR = BASE_DIR / "outputs"

MATRICES_DIR = OUTPUT_DIR / "matrices"
TABLAS_DIR = OUTPUT_DIR / "tablas"
FIGURAS_DIR = OUTPUT_DIR / "figuras"

TABLAS_DIR.mkdir(parents=True, exist_ok=True)
FIGURAS_DIR.mkdir(parents=True, exist_ok=True)


# =========================================================
# Cargar matrices de covarianza
# =========================================================

archivo_sigma_original = MATRICES_DIR / "sigma_muestral_80pct.csv"

archivo_sigma_limpia = MATRICES_DIR / "sigma_limpia_80pct.csv"

sigma_original = pd.read_csv(
    archivo_sigma_original,
    index_col=0
)

sigma_limpia = pd.read_csv(
    archivo_sigma_limpia,
    index_col=0
)

# Asegurar formato numérico
sigma_original = sigma_original.apply(
    pd.to_numeric,
    errors="coerce"
)

sigma_limpia = sigma_limpia.apply(
    pd.to_numeric,
    errors="coerce"
)

print("Dimensiones Sigma original:")
print(sigma_original.shape)

print("\nDimensiones Sigma limpia:")
print(sigma_limpia.shape)

display(sigma_original.head())
display(sigma_limpia.head())

# =========================================================
# Portafolio de mínima varianza
# Matriz original
# =========================================================

N = sigma_original.shape[0]

unos = np.ones(N)

# Inversa
sigma_original_inv = np.linalg.inv(
    sigma_original
)

# Pesos óptimos
pesos_original = (
    sigma_original_inv @ unos
)

pesos_original = (
    pesos_original
    /
    (unos.T @ sigma_original_inv @ unos)
)

print("Suma de pesos:")
print(pesos_original.sum())

print("\nPrimeros 10 pesos:")
print(pesos_original[:10])

# =========================================================
# Portafolio de mínima varianza
# Matriz limpia
# =========================================================

sigma_limpia_inv = np.linalg.inv(
    sigma_limpia
)

pesos_limpios = (
    sigma_limpia_inv @ unos
)

pesos_limpios = (
    pesos_limpios
    /
    (unos.T @ sigma_limpia_inv @ unos)
)

print("Suma de pesos:")
print(pesos_limpios.sum())

print("\nPrimeros 10 pesos:")
print(pesos_limpios[:10])

# =========================================================
# Comparación de pesos
# =========================================================

comparacion_pesos = pd.DataFrame({
    "Emisora": sigma_original.columns,
    "Peso_original": pesos_original,
    "Peso_limpio": pesos_limpios
})

comparacion_pesos["Diferencia"] = (
    comparacion_pesos["Peso_limpio"]
    -
    comparacion_pesos["Peso_original"]
)

display(
    comparacion_pesos.sort_values(
        "Diferencia",
        key=np.abs,
        ascending=False
    ).head(10)
)

# =========================================================
# Resumen de diferencias en pesos
# =========================================================

print(
    "Máxima diferencia absoluta:"
)

print(
    comparacion_pesos["Diferencia"]
    .abs()
    .max()
)

print(
    "\nPromedio diferencia absoluta:"
)

print(
    comparacion_pesos["Diferencia"]
    .abs()
    .mean()
)

#Guardado de la tabla de pesos por emisora
archivo_pesos = (
    TABLAS_DIR /
    "comparacion_pesos_markowitz.csv"
)

comparacion_pesos.to_csv(
    archivo_pesos,
    index=False
)

print(archivo_pesos)

# =========================================================
# Pesos óptimos de Markowitz
# Original vs Limpio
# =========================================================

plt.style.use("default")
plt.figure(figsize=(16,7))
facecolor="white"

x = np.arange(len(comparacion_pesos))

ancho = 0.40

plt.bar(
    x - ancho/2,
    comparacion_pesos["Peso_original"],
    width=ancho,
    color="#2F5597",
    label="Original"
)

plt.bar(
    x + ancho/2,
    comparacion_pesos["Peso_limpio"],
    width=ancho,
    color="#A64D79",
    label="Limpio RMT"
)

plt.axhline(
    0,
    color="black",
    linewidth=0.8
)

plt.title(
    "Pesos óptimos de Markowitz: matriz original vs matriz limpia",
    fontsize=14
)

plt.xlabel("Emisora")
plt.ylabel("Peso óptimo")

plt.xticks(
    x,
    comparacion_pesos["Emisora"],
    rotation=90,
    fontsize=8
)

plt.legend()

plt.grid(
    axis="y",
    alpha=0.25
)

plt.tight_layout()

archivo_barras_pesos = (
    FIGURAS_DIR /
    "pesos_markowitz_original_vs_limpio.png"
)

plt.savefig(
    archivo_barras_pesos,
    dpi=300,
    bbox_inches="tight"
)

plt.show()

print("Figura guardada en:")
print(archivo_barras_pesos)

# =========================================================
# Riesgo del portafolio de mínima varianza
# =========================================================

# Riesgo usando la matriz original
varianza_original = (
    pesos_original.T
    @ sigma_original.values
    @ pesos_original
)

riesgo_original = np.sqrt(varianza_original)

# Riesgo usando la matriz limpia
varianza_limpia = (
    pesos_limpios.T
    @ sigma_limpia.values
    @ pesos_limpios
)

riesgo_limpio = np.sqrt(varianza_limpia)

print("Varianza portafolio original:")
print(varianza_original)

print("\nRiesgo portafolio original:")
print(riesgo_original)

print("\nVarianza portafolio limpio:")
print(varianza_limpia)

print("\nRiesgo portafolio limpio:")
print(riesgo_limpio)

print("\nDiferencia en riesgo:")
print(riesgo_limpio - riesgo_original)

# =========================================================
# Cargar rendimientos logarítmicos
# =========================================================

DATOS_PROCESADOS_DIR = OUTPUT_DIR / "datos_procesados"

archivo_rendimientos = (
    DATOS_PROCESADOS_DIR /
    "rendimientos_logaritmicos_80pct.csv"
)

rendimientos = pd.read_csv(archivo_rendimientos)

rendimientos["Fecha"] = pd.to_datetime(
    rendimientos["Fecha"]
)

rendimientos = rendimientos.sort_values("Fecha")

R = rendimientos.drop(columns="Fecha")

R = R.apply(
    pd.to_numeric,
    errors="coerce"
)

# Asegurar mismo orden de emisoras que Sigma
R = R[sigma_original.columns]

print("Dimensiones de rendimientos:")
print(R.shape)

display(R.head())

# =========================================================
# Vector de rendimientos esperados
# =========================================================

mu = R.mean()

print("Primeros rendimientos promedio diarios:")
display(mu.head())

print("Dimensión de mu:")
print(mu.shape)

# =========================================================
# Función para calcular portafolio eficiente
# =========================================================

def portafolio_eficiente(sigma, mu, retorno_objetivo):
    """
    Calcula los pesos del portafolio eficiente
    para un rendimiento objetivo dado.

    sigma:
        Matriz de covarianza.

    mu:
        Vector de rendimientos esperados.

    retorno_objetivo:
        Rendimiento esperado objetivo del portafolio.
    """

    # Convertir a arreglos numéricos
    sigma_np = sigma.values
    mu_np = mu.values

    # Número de activos
    n = len(mu_np)

    # Vector de unos
    unos = np.ones(n)

    # Inversa de la matriz de covarianza
    sigma_inv = np.linalg.inv(sigma_np)

    # Constantes clásicas de Markowitz
    A = unos.T @ sigma_inv @ unos
    B = unos.T @ sigma_inv @ mu_np
    C = mu_np.T @ sigma_inv @ mu_np

    Delta = A * C - B**2

    # Multiplicadores
    lambda_1 = (C - B * retorno_objetivo) / Delta
    lambda_2 = (A * retorno_objetivo - B) / Delta

    # Pesos óptimos
    pesos = sigma_inv @ (
        lambda_1 * unos
        +
        lambda_2 * mu_np
    )

    return pesos

# =========================================================
# Retornos objetivo
# =========================================================

retornos_objetivo = np.linspace(mu.min(), mu.max(), 100)
emisora_min = mu.idxmin()
emisora_max = mu.idxmax()
retorno_min = mu.min()
retorno_max = mu.max()

print("Retorno mínimo:")
print(retorno_min)
print("Emisora:")
print(emisora_min)

print("\nRetorno máximo:")
print(retorno_max)
print("Emisora:")
print(emisora_max)

print("\nNúmero de puntos:")
print(len(retornos_objetivo))

# =========================================================
# Frontera eficiente
# Matriz original
# =========================================================

riesgos_original = []
retornos_original = []

for retorno_objetivo in retornos_objetivo:

    pesos = portafolio_eficiente(
        sigma_original,
        mu,
        retorno_objetivo
    )

    riesgo = np.sqrt(
        pesos.T
        @ sigma_original.values
        @ pesos
    )

    riesgos_original.append(
        riesgo
    )

    retornos_original.append(
        retorno_objetivo
    )

print(
    "Portafolios originales calculados:"
)

print(
    len(riesgos_original)
    )

# =========================================================
# Frontera eficiente
# Matriz limpia
# =========================================================

riesgos_limpios = []
retornos_limpios = []

for retorno_objetivo in retornos_objetivo:

    pesos = portafolio_eficiente(
        sigma_limpia,
        mu,
        retorno_objetivo
    )

    riesgo = np.sqrt(
        pesos.T
        @ sigma_limpia.values
        @ pesos
    )

    riesgos_limpios.append(
        riesgo
    )

    retornos_limpios.append(
        retorno_objetivo
    )

print(
    "Portafolios limpios calculados:"
)

print(
    len(riesgos_limpios)
)

# =========================================================
# Frontera eficiente
# Original vs Limpia
# =========================================================

plt.figure(figsize=(10,6))

plt.plot(riesgos_original, retornos_original, linewidth=2, label="Original")
plt.plot(riesgos_limpios, retornos_limpios, linewidth=2, label="Limpia RMT")
plt.title("Frontera eficiente: Original vs Limpia")
plt.xlabel("Riesgo")
plt.ylabel("Retorno esperado")
plt.legend()
plt.grid(True)
plt.tight_layout()

archivo_frontera = (FIGURAS_DIR /"frontera_eficiente_original_vs_rmt.png")

plt.savefig(archivo_frontera, dpi=300, bbox_inches="tight")
plt.show()

print("Figura guardada en:")
print(archivo_frontera)

# Riesgo mínimo de cada frontera

print(min(riesgos_original))
print(min(riesgos_limpios))

#Calculo de la variación porcentual
riesgo_min_original = min(riesgos_original)
riesgo_min_limpio = min(riesgos_limpios)
print((riesgo_min_original-riesgo_min_limpio)/riesgo_min_original * 100)

# =========================================================
# VALIDACIÓN IN-SAMPLE Y OUT-OF-SAMPLE
# =========================================================

# =========================================================
# Rango temporal de los datos
# =========================================================

print("Fecha inicial:")
print(rendimientos["Fecha"].min())
print("\nFecha final:")
print(rendimientos["Fecha"].max())

# =========================================================
# División temporal
# =========================================================

fecha_corte = "2022-01-01"

R_in = R[
    rendimientos["Fecha"] < fecha_corte
]

R_out = R[
    rendimientos["Fecha"] >= fecha_corte
]

print("Dimensiones in-sample:")
print(R_in.shape)

print("\nDimensiones out-of-sample:")
print(R_out.shape)

print("\nFecha inicial in-sample:")
print(rendimientos.loc[
    rendimientos["Fecha"] < fecha_corte,
    "Fecha"
].min())

print("\nFecha final in-sample:")
print(rendimientos.loc[
    rendimientos["Fecha"] < fecha_corte,
    "Fecha"
].max())

print("\nFecha inicial out-of-sample:")
print(rendimientos.loc[
    rendimientos["Fecha"] >= fecha_corte,
    "Fecha"
].min())

print("\nFecha final out-of-sample:")
print(rendimientos.loc[
    rendimientos["Fecha"] >= fecha_corte,
    "Fecha"
].max())

# =========================================================
# Matrices de covarianza
# In-Sample y Out-of-Sample
# =========================================================

sigma_in = R_in.cov()
sigma_out = R_out.cov()
print("Dimensiones Sigma in-sample:")
print(sigma_in.shape)
print("\nDimensiones Sigma out-of-sample:")
print(sigma_out.shape)

# =========================================================
# Limpieza RMT sobre la muestra de entrenamiento
# =========================================================

# =========================================================
# Matriz de correlación in-sample
# =========================================================

matriz_correlacion_in = R_in.corr()

print("Dimensiones matriz correlación:")
print(matriz_correlacion_in.shape)

# =========================================================
# Eigenvalores de correlación in-sample
# =========================================================

eigenvalores_in, eigenvectores_in = np.linalg.eigh(matriz_correlacion_in)
idx = np.argsort(eigenvalores_in)[::-1]
eigenvalores_in = eigenvalores_in[idx]
eigenvectores_in = eigenvectores_in[:, idx]
print("Primeros 10 eigenvalores:")
print(eigenvalores_in[:10])

# =========================================================
# Parámetro q
# =========================================================

N = R_in.shape[1]
T = R_in.shape[0]
q = N / T

print("N:")
print(N)

print("\nT:")
print(T)

print("\nq:")
print(q)

# =========================================================
# Límites Marcenko-Pastur
# =========================================================

lambda_inf = (
    1 - np.sqrt(q)
)**2

lambda_sup = (
    1 + np.sqrt(q)
)**2

print("Lambda inferior:")
print(lambda_inf)

print("\nLambda superior:")
print(lambda_sup)

# =========================================================
# Clasificación de eigenvalores
# =========================================================

eigenvalores_informativos = (
    eigenvalores_in[
        eigenvalores_in > lambda_sup
    ]
)

eigenvalores_ruido = (
    eigenvalores_in[
        eigenvalores_in <= lambda_sup
    ]
)

print(
    "Número eigenvalores informativos:"
)

print(
    len(eigenvalores_informativos)
)

print(
    "\nNúmero eigenvalores ruido:"
)

print(
    len(eigenvalores_ruido)
)

# =========================================================
# Eigenvalor promedio del ruido
# =========================================================

lambda_ruido = np.mean(
    eigenvalores_ruido
)

print(
    "Eigenvalor promedio del ruido:"
)

print(
    lambda_ruido
)

# =========================================================
# Eigenvalue clipping
# =========================================================

eigenvalores_clipped = np.copy(
    eigenvalores_in
)

eigenvalores_clipped[
    eigenvalores_clipped <= lambda_sup
] = lambda_ruido

print(
    "Primeros 10 eigenvalores originales:"
)

print(
    eigenvalores_in[:10]
)

print(
    "\nPrimeros 10 eigenvalores limpios:"
)

print(
    eigenvalores_clipped[:10]
)

# =========================================================
# Reconstrucción matriz correlación limpia
# =========================================================

correlacion_limpia_in = (
    eigenvectores_in
    @
    np.diag(eigenvalores_clipped)
    @
    eigenvectores_in.T
)

print(
    correlacion_limpia_in.shape
)

# =========================================================
# Renormalización diagonal
# =========================================================

D = np.sqrt(
    np.diag(correlacion_limpia_in)
)

correlacion_limpia_in = (
    correlacion_limpia_in
    /
    np.outer(D, D)
)

print(
    np.diag(correlacion_limpia_in)[:5]
)

# =========================================================
# Sigma limpia in-sample
# =========================================================

desv_est_in = np.sqrt(
    np.diag(sigma_in)
)

sigma_limpia_in = (
    np.diag(desv_est_in)
    @
    correlacion_limpia_in
    @
    np.diag(desv_est_in)
)

print(
    sigma_limpia_in.shape
)

# =========================================================
# Pesos óptimos con Sigma in-sample original
# =========================================================

N = sigma_in.shape[0]

unos = np.ones(N)

sigma_in_inv = np.linalg.inv(
    sigma_in.values
)

pesos_original_in = (
    sigma_in_inv @ unos
)

pesos_original_in = (
    pesos_original_in
    /
    (unos.T @ sigma_in_inv @ unos)
)

print(
    "Suma pesos original:"
)

print(
    pesos_original_in.sum()
)

# =========================================================
# Pesos óptimos con Sigma limpia in-sample
# =========================================================

sigma_limpia_in_inv = np.linalg.inv(
    sigma_limpia_in
)

pesos_rmt_in = (
    sigma_limpia_in_inv @ unos
)

pesos_rmt_in = (
    pesos_rmt_in
    /
    (unos.T @ sigma_limpia_in_inv @ unos)
)

print(
    "Suma pesos RMT:"
)

print(
    pesos_rmt_in.sum()
)

# =========================================================
# Riesgo in-sample
# =========================================================

riesgo_in_original = np.sqrt(
    pesos_original_in.T
    @ sigma_in.values
    @ pesos_original_in
)

riesgo_in_rmt = np.sqrt(
    pesos_rmt_in.T
    @ sigma_limpia_in
    @ pesos_rmt_in
)

print(
    "Riesgo in-sample original:"
)

print(
    riesgo_in_original
)

print(
    "\nRiesgo in-sample RMT:"
)

print(
    riesgo_in_rmt
)

# =========================================================
# Riesgo out-of-sample
# =========================================================

riesgo_out_original = np.sqrt(
    pesos_original_in.T
    @ sigma_out.values
    @ pesos_original_in
)

riesgo_out_rmt = np.sqrt(
    pesos_rmt_in.T
    @ sigma_out.values
    @ pesos_rmt_in
)

print(
    "Riesgo out-of-sample original:"
)

print(
    riesgo_out_original
)

print(
    "\nRiesgo out-of-sample RMT:"
)

print(
    riesgo_out_rmt
)

# =========================================================
# Resumen final
# =========================================================

resumen = pd.DataFrame({
    "Modelo": [
        "Original",
        "RMT"
    ],
    "Riesgo_in_sample": [
        riesgo_in_original,
        riesgo_in_rmt
    ],
    "Riesgo_out_sample": [
        riesgo_out_original,
        riesgo_out_rmt
    ]
})

display(
    resumen
)

resumen.to_csv(TABLAS_DIR /"comparacion_in_sample_out_sample.csv", index=False)

# =========================================================
# Tabla completa de pesos
# =========================================================

pesos_finales = pd.DataFrame({
    "Emisora": sigma_in.columns,
    "Peso_original": pesos_original_in,
    "Peso_RMT": pesos_rmt_in
})

pesos_finales["Diferencia"] = (
    pesos_finales["Peso_RMT"]
    -
    pesos_finales["Peso_original"]
)

display(
    pesos_finales.head()
)

archivo_pesos_finales = (
    TABLAS_DIR /
    "pesos_markowitz_in_sample.csv"
)

pesos_finales.to_csv(
    archivo_pesos_finales,
    index=False
)

print(
    archivo_pesos_finales
)

# =========================================================
# Resumen limpieza RMT
# =========================================================

resumen_rmt = pd.DataFrame({
    "Metrica": [
        "Activos",
        "Observaciones_in_sample",
        "q",
        "Lambda_inferior",
        "Lambda_superior",
        "Eigenvalores_informativos",
        "Eigenvalores_ruido",
        "Lambda_ruido"
    ],
    "Valor": [
        N,
        T,
        q,
        lambda_inf,
        lambda_sup,
        len(eigenvalores_informativos),
        len(eigenvalores_ruido),
        lambda_ruido
    ]
})

display(
    resumen_rmt
)

archivo_resumen_rmt = (
    TABLAS_DIR /
    "resumen_rmt.csv"
)

resumen_rmt.to_csv(
    archivo_resumen_rmt,
    index=False
)

print(
    archivo_resumen_rmt
)












```
